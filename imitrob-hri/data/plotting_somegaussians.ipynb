```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
import scipy
```

```python
def area_overlap_over_gaussians(mu1, sigma1, mu2, sigma2, erf=scipy.special.erf):
    c = (mu2 * sigma1 ** 2 - sigma2 * (mu1 * sigma2 + sigma1 * np.sqrt((mu1 - mu2) ** 2 + 2 * (sigma1 ** 2 - sigma2 ** 2) * np.log( sigma1 / sigma2))) ) / (sigma1 ** 2 - sigma2 ** 2)
    return 1 - erf((c - mu1) / (np.sqrt(2) * sigma1) ) / 2 + erf((c - mu2) / (np.sqrt(2) * sigma2) ) / 2
```

```python
def plot_gaussians(unsure_prob_mu, unsure_prob_sigma, activated_prob_mu, activated_prob_sigma):
    #x-axis ranges from -3 and 3 with .001 steps
    x = np.arange(-3, 3, 0.001)

    #plot normal distribution with mean 0 and standard deviation 1
    plt.plot(x, norm.pdf(x, unsure_prob_mu, unsure_prob_sigma))

    plt.plot(x, norm.pdf(x, activated_prob_mu, activated_prob_sigma))

    plt.title(f"Overlap: {round(area_overlap_over_gaussians(unsure_prob_mu, unsure_prob_sigma,activated_prob_mu, activated_prob_sigma)*100,4)}%")
```

Area overlap from two Gaussians tests

```python
area_overlap_over_gaussians(0.5, 0.2, 0.9, 0.2)
```

```python
mu1=5.28; sigma1=0.91
mu2=8.45; sigma2=1.36
area_overlap_over_gaussians(5.28, 0.91, 8.45, 1.36)
```

```python
plot_gaussians(unsure_prob_mu=0.5, unsure_prob_sigma=0.2, activated_prob_mu=0.9, activated_prob_sigma=0.21)
```
